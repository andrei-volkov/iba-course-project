### Calc specializations

```python
def get_all_specializations(data):
    temp = []

    for vac in data:
        if 'specializations' in vac:
            temp.extend(a['profarea_name'] for a in vac['specializations'])
    return set(temp)
```

```python
def get_new_specializations_features(data):
    all_spec = get_all_specializations(data)
    dataframe = pd.DataFrame()
    for spec in all_spec:
        lst = []
        for a in data:
            if 'specializations' in a:
                lst.append(sum([spec == s['profarea_name'] for s in a['specializations']]) > 0)
            else:
                lst.append(False)

        dataframe = dataframe.append(pd.Series(lst, name = spec))    
    return dataframe.T
```
